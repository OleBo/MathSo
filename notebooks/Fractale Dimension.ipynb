# Fractale Dimension
A [fractal dimension](https://en.wikipedia.org/wiki/Fractal_dimension) is a ratio providing a statistical index of complexity comparing how detail in a pattern (strictly speaking, a fractal pattern) changes with the scale at which it is measured.

# Box counting
[Box counting](https://en.wikipedia.org/wiki/Box_counting) is a method of gathering data for analyzing complex patterns by breaking a dataset, object, image, etc. into smaller and smaller pieces, typically "box"-shaped, and analyzing the pieces at each smaller scale.

## Examples
### Koch Curve
The [Koch curve](https://en.wikipedia.org/wiki/Koch_snowflake), was published in a 1904 paper titled "On a Continuous Curve Without Tangents, Constructible from Elementary Geometry" by the Swedish mathematician Helge von Koch. It is one of the earliest fractal curves to be described.

```python
sin60 = sqrt(3)/2

xdata = [0, 0.5, 1, 0]
ydata = [0, sin60, 0, 0]

fig, ax = plt.subplots()
ax.plot(xdata, ydata)
fig.tight_layout()
plt.show()
```

```python
def break_line(pos, xdata, ydata):
    """
    Breaks a line and form the next spike in a Koch iteration
    the line is represented by its ends x/ydata[pos,pos+1]
    """
    x0 = xdata[pos]
    x1 = xdata[pos+1]
    y0 = ydata[pos]
    y1 = ydata[pos+1]
    xvec = x1 - x0
    yvec = y1 - y0

    xdata.insert(pos+1, x0 + xvec/3)
    ydata.insert(pos+1, y0 + yvec/3)

    xdata.insert(pos+2, x0 + xvec/2 - yvec*sin60/3)
    ydata.insert(pos+2, y0 + xvec*sin60/3 + yvec/2)

    xdata.insert(pos+3, x0 + 2*xvec/3)
    ydata.insert(pos+3, y0 + 2*yvec/3)
```

```python
def koch_iteration(xdata, ydata):
    """
    Breaks the (n-1) lines in a xdata of length n by calling break_line()
    """
    pos = 0
    while pos < len(xdata) - 1:
        break_line(pos, xdata, ydata)
        pos += 4
```

```python
koch_iteration(xdata, ydata)
fig, ax = plt.subplots()
ax.plot(xdata, ydata)
fig.tight_layout()
plt.show()
```

Loop that creates each frame for animation

```python
for i in range(1, 7):
    koch_iteration(xdata, ydata)
    ax.plot(xdata, ydata)
fig, ax = plt.subplots()
ax.plot(xdata, ydata)
fig.tight_layout()
plt.grid(True)
plt.show()
```
