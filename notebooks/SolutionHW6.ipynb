Write a Python function xlsplot that loads an Excel table in Python and creates a simple two-dimensional plot from it. The table should contain two columns, the first with the x values, the second with the y values. The input argument of the function should be the name of the Excel file. The cell range to be plotted is to be passed as an optional argument. If the table contains column headings, these should be used to label the axes of the plot. The function should return a graphics handle to the plot. Create an example for using the function.

```python
!open scripts/xlsplot_example.xlsx
```

```python
import pandas as pd
def xlsplot(file, range=None):
    df = pd.read_excel(file)
    #range = kwargs.get('range', None)
    if range != None:
        df = df.loc[df['x'] >= range[0]]
        df = df.loc[df['x'] <= range[1]]
    return df.plot('x','y',title=file)
    
xlsplot('scripts/xlsplot_example.xlsx');
```
